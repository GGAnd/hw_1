```python
import nltk
nltk.download('punkt')

from nltk.tokenize import (
    sent_tokenize,
    word_tokenize,
    TweetTokenizer,
    WordPunctTokenizer,
    WhitespaceTokenizer,
    LegalitySyllableTokenizer,
    SyllableTokenizer,
    MWETokenizer,
    )
```

```python
words=word_tokenize(text)
words
```

```python
from spellchecker import SpellChecker
spell= SpellChecker(language='ru')
miss=spell.unknown(words)
for word in miss:
    corr=spell.correction(word)
    if corr:
      print(word)
      print(corr)
      print(spell.candidates(word))
      print('-'*20)
```

```python
sents=sent_tokenize(text,language='russian')
sents
```

```python
tokenizer= MWETokenizer();
tokenizer.add_mwe(('"Русское','пиво"'))
tokenizer.tokenize(text.split())[10:]
```

```python
morph = pymorphy2.MorphAnalyzer()
```

```python
for word in WordPunctTokenizer().tokenize(text):
   print(morph.parse(word))
   p = morph.parse(word)[0]
   print('tag:',p.tag)
   print('normal form:',p.normal_form)
   print('_'*20)
```

```python
from natasha import Doc,NewsEmbedding,NewsSyntaxParser,Segmenter,NewsMorphTagger, MorphVocab
```

```python
segmenter=Segmenter()
enb=NewsEmbedding()
morph_tagger=NewsMorphTagger(enb)
syntax_parser=NewsSyntaxParser(enb)
morph_vocad=MorphVocab()
```

```python
doc = Doc(dataset[0])
doc.segment(segmenter)
print(doc.sents[:5])
```

```python
doc.tag_morph(morph_tagger)
print(doc.tokens[:5])
```

```python
for token in doc.tokens:
    token.lemmatize(morph_vocad)
print(doc.tokens[:5])
```
